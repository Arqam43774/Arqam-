```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips = sns.load_dataset('tips')

# Count the number of people in the smoking area vs non-smoking area
smoking_area_count = tips['smoker'].value_counts()

# Display a pie chart
plt.figure(figsize=(7, 7))
plt.pie(smoking_area_count, labels=smoking_area_count.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99'])
plt.title('Percentage of People in Smoking Area vs Non-Smoking Area')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```
